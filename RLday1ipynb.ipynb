```python
import gym
```

```python
#Create and load the simulation environment

env = gym.make("CartPole-v1")
```

```python
# Environment : CartPole
# Documentation Link: https://www.gymlibrary.dev/environments/classic_control/cart_pole/
# Goal is to balance the pole on the cart by moving the cart left or right
#
# Agent: Cart
#
# Actions: 0 --- Left
#          1 --- Right
#
# State: ["Cart Position", "Cart Velocity", "pole Angle", "poleVelocity"]
#
# Reward: 1 for every step taken
#
# Termination Condition:
# 1. Termination: Pole Angle is greater than ±12°
# 2. Termination: Cart Position is greater than ±2.4 (center of the cart reaches the edge of the display)
# 3. Truncation: Episode length is greater than 500 (200 for v0)
```

# Common Functions used in OpenAI Gym

```python
# How to get the initial state?
env.reset()
```

```python
# How to generate random action?
action = env.action_space.sample()
action
```

```python
# How to supply the action to the environemnt?
# --- or ----
# How to execute an action

env.step(action)


#step() accepts action and returns a tuple containing the following information:
#a. nextState
#b. reward
#c. isTerminated Boolean
#d. debugInfo 
```

# Run a single episode

```python
# initialize the observation/state
env.reset()

for episodeSteps in range(500):
    # choose a random action
    action = env.action_space.sample()

    # Supply action to the env
    newState, reward, isTerminated, debugInfo = env.step(action)

    # Print info
    print("Episode Step {} Given Action {} I got reward {} and the next state is {}".format(
        episodeSteps, action, reward, newState))

    # Check for termination
    if isTerminated:
        print("Terminated !!!")
        break

env.close()
```

# Run a multiple episode

```python
# run 10 episodes

for episodeCount in range(1, 11):
    env.reset()
    print("\nEpisode No:", episodeCount)

    for episodeSteps in range(500):
        # choose a random action
        action = env.action_space.sample()

        # Supply action to the env
        newState, reward, isTerminated, debugInfo = env.step(action)

        # Print info
        print("Episode Step {} Given Action {} I got reward {} and the next state is {}".format(
            episodeSteps, action, reward, newState))

        # Check for termination
        if isTerminated:
            print("Terminated!!!")
            break

env.close()
```
